```python
import polars as pl
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pl.read_csv('Stacked_Employer_data.csv')
data.head()
```

```python
data_means = (data
              .group_by(pl.col('Industry','Major'))
              .agg(pl.col('Likert Score').mean().alias('Mean Score'))
             )
data_means.head()
```

```python
from plotnine import ggplot, geom_point, aes

(ggplot(data = data_means) 
+ geom_point(aes(x = "Major",
                  y = "Mean Score"),
                  color = 'Industry',
                  size = 6)
 
)
```
